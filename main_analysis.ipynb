```python
# Main analysis notebook
import sys
sys.path.append('..')

from data.preprocessing import *
from data.analysis import *
from models.state_space import *
from models.rl_model import *
from visualization.plots import *
from config import settings

# Load and preprocess data
print("Loading data...")
df = load_data()
print(f"Original shape: {df.shape}")

# Clean data
print("Cleaning numeric data...")
df_clean = clean_numeric_data(df)
df_clean = remove_all_null_rows(df_clean)
print(f"After cleaning: {df_clean.shape}")

# Basic statistics
print("\nBasic statistics:")
print(f"Unique patients: {len(df_clean['ID'].unique())}")
missing_info = get_missingness_summary(df_clean)
print(missing_info.head())

# Lab test completeness analysis
print("\nAnalyzing lab test completeness...")
table_s1 = analyze_lab_test_completeness(df_clean)
print(table_s1.head())

# Time point analysis
print("\nTime point analysis:")
time_summary = analyze_time_points(df_clean)
for key, value in time_summary.items():
    print(f"{key}: {value:.2f}")

# Build dynamic states
print("\nBuilding dynamic states...")
df_y = build_dynamic_states(df_clean)
print(f"States built: {df_y.shape}")

# Prepare RL data
print("\nPreparing RL data...")
df_rl = prepare_rl_data(df_y)
print(f"RL data shape: {df_rl.shape}")
print(df_rl.head())

# Train RL model
print("\nTraining RL model...")
result = train_rl_model(df_rl)
print("Training complete!")

# Visualizations
print("\nGenerating visualizations...")

# State distribution
df_counts = df_y.groupby(["Y_x", "Y_y"]).size().reset_index(name="count")
fig1 = plot_state_distribution(df_counts)
plt.show()

# Treatment distribution
fig2 = plot_treatment_distribution(df_y)
plt.show()

# Reward distribution
fig3 = plot_reward_distribution(df_rl)
plt.show()

# Training results
# (Need to compute per_patient_df first - would add helper function)
# fig4 = plot_training_results(result, per_patient_df)
# plt.show()

print("\nAnalysis complete!")
```
